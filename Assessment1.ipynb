```python
#import libraries

import pandas as pd
import numpy as np
```

```python
bank = pd.read_excel('Bank.xlsx')
account = pd.read_excel('Account.xlsx')
customer = pd.read_excel('Customer.xlsx')
loan = pd.read_excel('Loan.xlsx')
transaction = pd.read_excel('Transaction.xlsx')
```

```python
account_type = (account[['Type']]
               .drop_duplicates()
                .reset_index(drop=True)
                .assign(account_type_id=lambda x: range(1, len(x)+1))
                [['account_type_id', 'Type']]
               )
```

```python
account_type
```

```python
# Joining 'account_type_id' to 'account' DataFrame based on 'Type'

account_normalized = pd.merge(account, account_type, on='Type', how='left')
```

```python
account_normalized
```

```python
account_normalized = account_normalized.drop(['Type'], axis=1)
```

```python
account_normalized.set_index('Account_NO')
```

```python
account_normalized.to_csv('account_n.csv')
account_type.to_csv('account_type.csv')
```

```python
loan_type = (loan[['Loan_Type']]
             .drop_duplicates()
             .reset_index(drop=True)
             .assign(loan_type_id=lambda x: range(1, len(x)+1))
             [['loan_type_id', 'Loan_Type']]

)
```

```python
loan_type
```

```python
#joining loan_type_id to loan Dataframe on loan_type

loan_normalized = pd.merge(loan, loan_type, on='Loan_Type', how='left')
```

```python
loan_normalized
```

```python
loan_normalized = loan_normalized.drop(['Loan_Type'], axis=1)
loan_normalized.set_index('Loan_ID')
```

```python
loan_normalized.to_csv('loan_n.csv')
```
