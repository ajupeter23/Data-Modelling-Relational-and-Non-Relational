```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Uploaling Bank dataset for normalization
df = pd.read_csv('Bank_Data.csv')
```

```python
# Display the Bank data
df.head()
```

#### In the given bank data, we can normalize it into five entities,1-Bank, 2-Account, 3-Customer, 4-Loan, and 5-Transaction.

```python
# Creating Bank entity
bank_columns = ['Branch_ID', 'Bank_Name', 'Bank_Contact', 'Bank_Location']
banks = df[bank_columns].drop_duplicates()

# Creating Customer entity
customer_columns = ['Customer_ID', 'Customer_Name', 'Customer_Contact', 'Customer_Email', 'Customer_Address']
customers = df[customer_columns].drop_duplicates()

# Creating Account entity
account_columns = ['Account_NO', 'Customer_ID', 'Account_Type', 'Account_Balance']
accounts = df[account_columns].drop_duplicates()

# Creating Transaction entity
transaction_columns = ['Transaction_ID', 'Account_NO', 'Tran_Date', 'Deposite', 'Withdrawal']
transactions = df[transaction_columns].drop_duplicates()

# Creating Loan entity
loan_columns = ['Loan_ID', 'Customer_ID', 'Loan_Type', 'Loan_Amount']
loans = df[loan_columns].drop_duplicates()
```

```python
# Display "banks"
banks = banks.set_index(['Branch_ID'])
banks
```

```python
# Display "customers"
customers = customers.set_index(['Customer_ID'])
customers
```

```python
# Display "accounts"
accounts = accounts.set_index(['Account_NO'])
accounts
```

```python
# Dropping null values
loans = loans.dropna()
```

```python
loans = loans.set_index(['Loan_ID'])
# Display "loans"
loans
```

```python
# Dropping null values
transactions = transactions.dropna()
```

```python
transactions = transactions.set_index(['Transaction_ID'])

# Display "transactions"
transactions
```

```python
transactions.to_csv('Transaction_N.csv')
loans.to_csv('Loan_N.csv')
accounts.to_csv('Account_N.csv')
customers.to_csv('Customer_N.csv')
banks.to_csv('Bank_N.csv')
```
